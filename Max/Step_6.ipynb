## Step 6

We use forecasts from the different regression models to bid on the Day-Ahead market. We calculate the revenues of the day ahead market and the balancing market using the realisation of the wind power, and assess if the model was good or not by comparing the output revenue with the revenues when the forecasts are equal to the realisation.

```python
# load the forecasted data
prediction = pd.read_csv('data/Wind Production Forecast.csv')

# load the prediction you need to evaluate
```

```python
# Load the data
merged_data = pd.read_csv('data/merged_data.csv')

# convert 'ts' column to datetime format
data_opti = merged_data.copy()
data_opti['ts'] = pd.to_datetime(merged_data['ts'])

# create a column with a day counter, knowing that ts is in the format 2023-01-02 00:00:00
data_opti['day_counter'] = (data_opti['ts'] - data_opti['ts'][0]).dt.days

# filter merged_data to have only the days higher than day_init and the days lower than day_end
day_init = 5
day_end = 11
```

```python
# create a dictionary to store the results
results_prediction = {}


# Run the optimization for the defined period
for day in range(day_init, day_end):
    print('Optimizing day', day)
    data_opti_day = data_opti[data_opti['day_counter'] == day]
    wind_forecast = {str(i): prediction.iloc[i, 1] for i in range(24)}
    input_data = InputData(wind_forecast, data_opti_day)

    step_6_optimized_bid = Step_6_Optimized_Bid(input_data)
    step_6_optimized_bid.run()
    step_6_optimized_bid.display_results()

    # Save the results
    results_prediction[day] = sum(step_6_optimized_bid.results.total_realised_profit[str(i)] for i in range(24))

# print the results
print(results_prediction)




```
